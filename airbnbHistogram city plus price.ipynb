```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
dataFrame = pd.read_csv("./Aemf1.csv")
print(dataFrame)
```

```python
average_price_city = dataFrame.groupby("City")['Price'].mean()
print(average_price_city)

```

```python
plt.bar(average_price_city.index, average_price_city.values, color='blue', alpha=0.7)
plt.title('Average Prices by City')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
