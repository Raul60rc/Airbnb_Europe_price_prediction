```python
mapeado = {False: 0, True: 1}

df["Shared Room"]=df["Shared Room"].map(mapeado)
df["Private Room"]=df["Private Room"].map(mapeado)
df["Superhost"]=df["Superhost"].map(mapeado)
```

       'Paris'= 6 // 'Rome'= 7//'Vienna'= 8

```python
cities = {'Amsterdam': 0, 'Athens' : 1, 'Barcelona': 2 , 'Berlin':3 , 'Budapest' : 4, 'Lisbon': 5, 'Paris': 6 , 'Rome': 7,'Vienna': 8}
df["City"] = df["City"].map(cities)
```

```python
df["Day"].unique()
```

'Weekday' = 0 // 'Weekend' = 1

```python
days = {'Weekday' : 0, 'Weekend': 1}
df["Day"] = df["Day"].map(days)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
df = df.drop(columns=["City","Day","Shared Room","Superhost","Multiple Rooms","Business","Cleanliness Rating","Guest Satisfaction","City Center (km)","Metro Distance (km)","Restraunt Index"])
```

```python
df
```

```python
x = df[["Room Type","Private Room","Person Capacity","Bedrooms","Attraction Index","Normalised Attraction Index","Normalised Restraunt Index"]]
y = df["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=33)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)
print("Error cuadrado medio (MSE):", mse)
```
